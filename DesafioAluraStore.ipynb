### Importa√ß√£o dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

```python
# C√âLULA 1: PREPARA√á√ÉO ESSENCIAL DE DADOS
import pandas as pd

# Nomes corretos das colunas
colunas_corretas = ['Produto', 'Categoria do Produto', 'Pre√ßo', 'Frete', 'Data da Compra', 'Vendedor', 'Local da compra', 'Avalia√ß√£o da compra', 'Tipo de pagamento', 'Quantidade de parcelas', 'lat', 'lon']

# Renomear colunas e adicionar o identificador de loja
loja.columns = colunas_corretas
loja['Loja'] = 'Loja_1'
loja2.columns = colunas_corretas
loja2['Loja'] = 'Loja_2'
loja3.columns = colunas_corretas
loja3['Loja'] = 'Loja_3'
loja4.columns = colunas_corretas
loja4['Loja'] = 'Loja_4'

# Unificar todos os DataFrames
df_comparativo = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)

# Criar a coluna de Receita Total
df_comparativo['Receita Total'] = df_comparativo['Pre√ßo'] + df_comparativo['Frete']

print("Dados de todas as lojas unificados em 'df_comparativo' com sucesso!")
```

#1. An√°lise do faturamento



```python
# C√âLULA 2: AN√ÅLISE DA M√âTRICA 1 (Faturamento e Ticket M√©dio)

# Fun√ß√£o de formata√ß√£o para Real Brasileiro
def formatar_brl(valor):
    if pd.isna(valor):
        return '-'
    return f"R$ {valor:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")

# Agrupar por 'Loja' e calcular o Faturamento Total e o Ticket M√©dio
faturamento_por_loja = df_comparativo.groupby('Loja').agg(
    Faturamento_Total=('Receita Total', 'sum'),
    Ticket_Medio=('Receita Total', 'mean'),
    Total_de_Vendas=('Produto', 'count')
).sort_values(by='Faturamento_Total', ascending=False)

# Aplicar formata√ß√£o
faturamento_por_loja['Faturamento_Total'] = faturamento_por_loja['Faturamento_Total'].apply(formatar_brl)
faturamento_por_loja['Ticket_Medio'] = faturamento_por_loja['Ticket_Medio'].apply(formatar_brl)

# Imprimir o resultado
print("üí∞ M√©trica 1: Comparativo de Faturamento e Ticket M√©dio por Loja")
print("---------------------------------------------------------------")
print(faturamento_por_loja)
```

# 2. Vendas por Categoria

```python

# C√âLULA 3: AN√ÅLISE DA M√âTRICA 2 (Vendas por Categoria)

# Agrupar por Loja e por Categoria do Produto, calculando a Receita Total
vendas_por_categoria_comparativo = df_comparativo.groupby(['Loja', 'Categoria do Produto']).agg(
    Receita_Total=('Receita Total', 'sum'),
    Total_de_Vendas=('Produto', 'count')
).reset_index()

# Ordenar por Loja e, em seguida, pela Receita Total (decrescente) para ver as top categorias de cada loja
vendas_por_categoria_comparativo = vendas_por_categoria_comparativo.sort_values(
    by=['Loja', 'Receita_Total'], ascending=[True, False]
)

# Aplicar a fun√ß√£o de formata√ß√£o (assumindo que 'formatar_brl' foi definida anteriormente)
vendas_por_categoria_comparativo['Receita_Total'] = vendas_por_categoria_comparativo['Receita_Total'].apply(formatar_brl)

# Imprimir o resultado
print("üõçÔ∏è M√©trica 2: Comparativo de Vendas por Categoria (Receita e Quantidade)")
print("----------------------------------------------------------------------")
print(vendas_por_categoria_comparativo)
```

# 3. M√©dia de Avalia√ß√£o das Lojas


```python
# C√âLULA 4: AN√ÅLISE DA M√âTRICA 3 (M√©dia de Avalia√ß√£o das Lojas)

# Agrupar por 'Loja' e calcular a m√©dia da Avalia√ß√£o
avaliacao_por_loja = df_comparativo.groupby('Loja').agg(
    Media_Avaliacao=('Avalia√ß√£o da compra', 'mean'),
    Total_Avaliacoes=('Avalia√ß√£o da compra', 'count')
).sort_values(by='Media_Avaliacao', ascending=False)

# Aplicar formata√ß√£o para duas casas decimais
avaliacao_por_loja['Media_Avaliacao'] = avaliacao_por_loja['Media_Avaliacao'].apply(lambda x: f'{x:.2f}')

# Imprimir o resultado
print("‚≠ê M√©trica 3: Comparativo da M√©dia de Avalia√ß√£o das Lojas")
print("----------------------------------------------------------")
print(avaliacao_por_loja)
```

# 4. Produtos Mais e Menos Vendidos

```python
# C√âLULA 5: AN√ÅLISE DA M√âTRICA 4 (Produtos Mais e Menos Vendidos)

print("üì¶ M√©trica 4: Produtos Mais e Menos Vendidos (por Quantidade)")
print("----------------------------------------------------------------------")

# Loop para analisar cada loja separadamente
for loja_nome in df_comparativo['Loja'].unique():
    # Filtra o DataFrame para a loja atual
    df_loja = df_comparativo[df_comparativo['Loja'] == loja_nome]

    # Conta a frequ√™ncia de cada produto na loja
    vendas_por_produto_loja = df_loja['Produto'].value_counts()

    # ----------------------------------------------------
    # Top 3 Mais Vendidos
    top_3_mais_vendidos = vendas_por_produto_loja.head(3)

    # ----------------------------------------------------
    # Top 3 Menos Vendidos (Produtos com menor volume, pode ser 1 venda)
    # Ordena pelo volume crescente e pega os 3 primeiros
    top_3_menos_vendidos = vendas_por_produto_loja.tail(3).sort_values(ascending=True)

    # Imprimir os resultados para a loja atual
    print(f"\n--- {loja_nome} ---")

    print("\n   üìà Top 3 Produtos Mais Vendidos:")
    print(top_3_mais_vendidos)

    print("\n   üìâ Top 3 Produtos Menos Vendidos:")
    print(top_3_menos_vendidos)

print("\n----------------------------------------------------------------------")
```

# 5. Frete M√©dio por Loja

```python
# C√âLULA 6: AN√ÅLISE DA M√âTRICA 5 (Frete M√©dio por Loja)

# Agrupar por 'Loja' e calcular a m√©dia do Frete
frete_por_loja = df_comparativo.groupby('Loja')['Frete'].mean().reset_index()
frete_por_loja.rename(columns={'Frete': 'Frete_Medio'}, inplace=True)
frete_por_loja = frete_por_loja.sort_values(by='Frete_Medio', ascending=False)

# Aplicar a fun√ß√£o de formata√ß√£o (assumindo que 'formatar_brl' foi definida anteriormente)
frete_por_loja['Frete_Medio'] = frete_por_loja['Frete_Medio'].apply(formatar_brl)

# Imprimir o resultado
print("üöö M√©trica 5: Comparativo do Frete M√©dio por Loja")
print("--------------------------------------------------")
print(frete_por_loja)
```

# 6. Resumo e Conclu√ß√£o: Tabela Resumo Final de Desempenho

```python
# C√âLULA 7: TABELA RESUMO FINAL DE DESEMPENHO

# 1. Agrupar e calcular as m√©tricas chaves
tabela_resumo = df_comparativo.groupby('Loja').agg(
    Faturamento_Total=('Receita Total', 'sum'),
    Ticket_Medio=('Receita Total', 'mean'),
    Avaliacao_Media=('Avalia√ß√£o da compra', 'mean'),
    Frete_Medio=('Frete', 'mean'),
    Total_de_Vendas=('Produto', 'count')
).sort_values(by='Faturamento_Total', ascending=False)

# 2. Aplicar formata√ß√£o para melhor visualiza√ß√£o
tabela_resumo['Faturamento_Total'] = tabela_resumo['Faturamento_Total'].apply(formatar_brl)
tabela_resumo['Ticket_Medio'] = tabela_resumo['Ticket_Medio'].apply(formatar_brl)
tabela_resumo['Frete_Medio'] = tabela_resumo['Frete_Medio'].apply(formatar_brl)
tabela_resumo['Avaliacao_Media'] = tabela_resumo['Avaliacao_Media'].apply(lambda x: f'{x:.2f}')

# 3. Imprimir a tabela e a recomenda√ß√£o

print("===============================================================")
print("  RESUMO COMPARATIVO FINAL PARA DECIS√ÉO DE VENDA (5 M√©tricas)")
print("===============================================================")
print(tabela_resumo)
print("===============================================================")

# 4. Gera√ß√£o da Recomenda√ß√£o de Venda (Baseada em Desempenho)

# Encontra a loja com o menor faturamento
pior_faturamento = df_comparativo.groupby('Loja')['Receita Total'].sum().idxmin()

# Encontra a loja com a pior avalia√ß√£o
pior_avaliacao = df_comparativo.groupby('Loja')['Avalia√ß√£o da compra'].mean().idxmin()

# Encontra a loja com o frete m√©dio mais alto (maior custo)
maior_custo_frete = df_comparativo.groupby('Loja')['Frete'].mean().idxmax()


print("\nüí° RECOMENDA√á√ÉO BASEADA NOS DADOS:")
print("-----------------------------------")

if pior_faturamento == pior_avaliacao:
    print(f"A **{pior_faturamento}** √© a principal candidata √† venda.")
    print(f"Motivo: Ela apresenta o pior faturamento total e a pior m√©dia de avalia√ß√£o de clientes, combinando baixo retorno financeiro e alto risco de reputa√ß√£o.")
elif pior_faturamento == maior_custo_frete:
    print(f"A **{pior_faturamento}** √© a principal candidata √† venda.")
    print(f"Motivo: Ela apresenta o pior faturamento total e o maior custo de frete, indicando baixo retorno financeiro e alta inefici√™ncia log√≠stica.")
else:
    print(f"A loja com o **Pior Desempenho Financeiro** √© a **{pior_faturamento}**.")
    print(f"A loja com a **Pior Satisfa√ß√£o do Cliente** √© a **{pior_avaliacao}**.")
    print(f"\nRecomenda√ß√£o: Avaliar se a **{pior_faturamento}** (menor faturamento) √© a melhor op√ß√£o, a menos que a {pior_avaliacao} tenha custos de revers√£o de reputa√ß√£o insustent√°veis.")

print("-----------------------------------")
```
